### Key Data Pipeline Processes

#### Stages 
1. Data Extraction
2. Data Ingestion
3. Transformation Stages (optional)
4. Loading of data into destination facility
5. Moitoring Workflows
6. Maintenance and optimization

#### Pipeline monitoring considerations
1. Latency
2. Throughput
3. Warnings, errors, failures - e.g. network overloading, destination failure
4. Utilization rate - how full piepeline resources are being utilized
5. Logging and alerting system

#### Load Balanced Pipelines
* Just-in-time data packet relays
* No upstream data flow bottlenecks
* Uniform packet throughput for each stage - such a pipeline is balanced

#### Handling Unbalanced Loads

##### Parallelization



- Pipelines typically contain bottlenecks
- Slower stages may be parallelized to speed up throughput
- Processes can be replicated on multiple CPUs, cores, threads
- Data packets are then distributed acrosse these channels
- Such pipelines are called dynamic or non-linear

#### Stage Sybchronization
- I/O buffers can help synchronize stages
- Holding area for data between processing stages
- Buffers regulate the flow of data, may improve throughput
- I/O buffers used to distrbute loads on parallelized stages
