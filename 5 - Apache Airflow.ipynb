### Apache Airflow Overview

- open source workflow orchestration tool
- used to programatically author, schedule, and monitor workflows
- a platform that lets you build and run workflows
- workflow is represented as a DAG
- not a data streaming solution
- used commonly for machine learning pipeline dependencies

### Advantages of Using Data Pipelines as DAGs in Apache Airflow

- DAG is a special kind of graph
- Directed graph - means that each edge has a specified direction
- Acyclic graph - has no loops/cycles or sequences of directed edges that return to a node in a chain (red)



#### Examples of DAGs




- DAG represents workflows
- Each task in a pipeline is represented by a node in a DAG
- Edges are dependencies
- Tasks run in order of the graph

#### Tasks and Operators
- Tasks are written in Python
- Tasks implement operators for example: Python, SQL, or Bash scripts
- Operators determine what each task does.
- Sensor operators poll for a certain time or condition
